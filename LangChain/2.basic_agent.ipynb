```python
from langchain_core.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.prebuilt import create_react_agent
from dotenv import load_dotenv
import os
```

```python
# Set your key
load_dotenv()

os.environ['GOOGLE_API_KEY'] = os.getenv('GOOGLE_API_KEY')
```

```python
# 1. Define the tool using the @tool decorator
@tool
def get_weather(city: str) -> str:
    '''
    Get weather for a given city.
    '''
    return f"It's always sunny in {city}!"


# 2. Initialize the Gemini model
model = ChatGoogleGenerativeAI(model='gemini-2.5-flash')

# 3. Create the agent
# In the modern syntax, the 'tools' and 'model' are passed directly
tools = [get_weather]
agent = create_react_agent(model, tools)

# 4. Run the agent
# LangGraph agents use a "state" dictionary (usually a list of messages)

response = agent.invoke({
    'messages': [('user', 'what is the weather in Lagos')]
})

# Print the last message from the response
print(response['messages'][-1].content)
```
